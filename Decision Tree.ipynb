```python
data=pd.read_csv('wns_train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data['education'].unique()
```

```python
data
```

```python
data['previous_year_rating']=data['previous_year_rating'].fillna(3)
```

```python
data['education']=data['education'].fillna("Bachelor's")
```

```python
data['education'].unique()
```

```python
data['education'].describe()
```

```python
out=data['is_promoted']
```

```python
inp=data.drop(['is_promoted','employee_id'],axis=1)
```

```python
data_dum=pd.get_dummies(inp,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(data_dum,out,test_size=0.3,random_state=48)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(xtrain,ytrain)
```

```python
ypred=log_model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc=accuracy_score(ytest,ypred)
acc
```

```python
con=confusion_matrix(ytest,ypred)
con
```

```python
cr=classification_report(ytest,ypred)
print(cr)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree_mod1=DecisionTreeClassifier()
dtree_mod1.fit(xtrain,ytrain)
```

```python
ypred=dtree_mod1.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc1=accuracy_score(ytest,ypred)
acc1
```

```python
cr1=classification_report(ytest,ypred)
print(cr1)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
folds=KFold(n_splits=5,shuffle=True,random_state=48)
hyper_params=[{"criterion": ['gini', 'entropy']}]
```

```python
mod=GridSearchCV(rdf_mod,param_grid=hyper_params,scoring="f1",cv=folds)
```

```python
mod.fit(xtrain,ytrain)
```

```python
mod.best_params_
```

```python
mod1= RandomForestClassifier(criterion = 'entropy')
```

```python
hyper_params=[{"max_depth": [10,20,30,40,50]}]
mod2=GridSearchCV(mod1,param_grid=hyper_params,scoring="f1",cv=folds)
mod2.fit(xtrain,ytrain)
```

```python
mod2.best_params_
```

```python
mod3 = RandomForestClassifier(criterion='entropy',max_depth=30)
hyper_params=[{'max_features':['auto','sqrt','log2']}]
mod4=GridSearchCV(mod3,param_grid=hyper_params,scoring="f1",cv=folds)
pred=mod4.fit(xtrain,ytrain)
```
